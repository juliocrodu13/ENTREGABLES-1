```python
# 1. Carga del Dataset:
# Carga el dataset proporcionado sobre películas en un DataFrame de pandas.

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv('/content/movie_metadata.csv')
```

```python
# 2. Revisión General de los Datos:

# Imprime las primeras 5 filas del dataset.

dataset.head()

# Usa .info() y .describe() para entender mejor la estructura y las estadísticas básicas.

dataset.info()
dataset.describe()
```

```python
# 3. Búsqueda y Eliminación de Duplicados:

# Verifica si hay filas duplicadas en el dataset utilizando el método .duplicated().

dataset.duplicated().sum()

print("Hay" , dataset.duplicated().sum() , " datos duplicados.")
```

```python
# Si se encuentran duplicados, elimínalos con .drop_duplicates() y verifica cuántos duplicados fueron eliminados.

dataset.drop_duplicates(inplace = True)
dataset.duplicated().sum()


print("Hay" , dataset.duplicated().sum() , " datos duplicados.")

```

```python
# 4. Identificación de Valores Nulos:

dataset.isnull().sum()

# Identifica qué columnas tienen valores nulos y cuántos valores nulos hay en cada columna.

```

```python
# 5. Visualización de Variables:

import matplotlib.pyplot as plt
import seaborn as sns

# Crea un histograma que muestre la distribución de la variable imdb_score.


plt.hist(dataset['imdb_score'], bins = 20, edgecolor = 'black', color = 'pink')
plt.title('Distribución de puntajes IMDB')
plt.xlabel('Puntaje IMDB')
plt.ylabel('Frecuencia')
plt.show()


# Realiza un gráfico de barras para visualizar la cantidad de películas por género (genres).

dataset['genres'].value_counts().head(20).plot(kind = "bar", edgecolor = "black", color = "pink")
plt.title("Géneros de películas: Top 20")
plt.xlabel('Género')
plt.ylabel('Cantidad de películas')
plt.show()
```

```python
# 6. Análisis de Correlaciones:

# Calcula y muestra la correlación entre budget, gross y imdb_score.

dataset["budget"] = dataset["budget"].str.replace("$", '', regex = False)
dataset_mean = dataset.copy ()
dataset_mean["budget"].isnull().sum()
```

```python
correlation = dataset[["budget", "gross", "imdb_score"]].corr()
correlation
```

```python
# Usa un mapa de calor para visualizar las correlaciones.

mapacalor = sns.heatmap(correlation, annot = True, cmap = "pink")
plt.show()
```

```python
# 7. Gráficos de Dispersión:

# Crea un gráfico de dispersión para analizar la relación entre budget y gross.

dispersion = plt.scatter(dataset["budget"], dataset["gross"], c = "pink")
plt.xlabel("Presupuesto")
plt.ylabel("Recaudación")
plt.title("Relación entre presupuesto y recaudación")
plt.show()

# Haz un gráfico de dispersión entre imdb_score y budget.

dispersion = plt.scatter(dataset["budget"], dataset["imdb_score"], c = "magenta")
plt.xlabel("Presupuesto")
plt.ylabel("Puntaje IMDB")
plt.title("Relación entre presupuesto y puntaje IMDB")
plt.show()
```

# Conclusiones
- La mayoría de las películas cuentan con un puntaje IMDB en el rango de 6 a 7.
- Las puntuaciones muy bajas o muy altas son escasas.
- Los géneros más frecuentes son Drama, Comedia y Thriller, mientras que Western, Musical y Bélico
se encuentran entre los menos comunes.
- Hay una correlación positiva entre el presupuesto de una película y su recaudación,
lo que indica que las producciones con mayor inversión tienden a generar ingresos más altos.
- La correlación entre el presupuesto y el puntaje IMDB es débil, lo que significa que una
mayor inversión no garantiza una mejor calificación.
- El puntaje IMDB de una película se ve afectado por varios factores, como el género, el presupuesto y la recaudación.
- De todos los géneros, el Drama es el más popular, seguido de Comedia y Thriller.
- Aunque las películas con mayor presupuesto suelen recaudar más, esto no asegura que obtengan mejores
calificaciones en IMDB.
